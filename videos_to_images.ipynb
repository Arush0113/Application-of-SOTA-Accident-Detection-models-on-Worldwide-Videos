```python
# all_videos_path = Config.TRAIN_VIDEOS_PATH
all_videos_path = ["./data/video/single_video_test"]

def create_frames_from_videos():
    for data_file in all_videos_path:
        print("Current Data Path: {} ".format(data_file))
        for f in sorted(listdir(data_file)):
            print("Current Video File: {} ".format(join(data_file, f)))
            
            cam = cv2.VideoCapture(join(data_file, f))

            # try:

            #     # creating a folder named data
            #     if not os.path.exists('data'):
            #         os.makedirs('data')

            # # if not created then raise error
            # except OSError:
            #     print ('Error: Creating directory of data')

            # frame
            currentframe = 0
            
#             try:
            os.makedirs(join('./data/images/AICITY_US_train_data', f))
#             break
#             except OSError, e:
#                 if e.errno != errno.EEXIST:
#                     raise   
#                 # time.sleep might help here
#                 pass

            while(True):

                # reading from frame
                ret,frame = cam.read()
#                 os.mkdir(join(data_file, f))
                
                

                if ret:
                    # if video is still left continue creating images
                    
                    name = str(join('./data/images/AICITY_US_train_data', f)) + '/' + 'frame' + str(currentframe) + '.jpg'
                    print ('Creating...' + name)

                    # writing the extracted images
                    cv2.imwrite(name, frame)

                    # increasing counter so that it will
                    # show how many frames are created
                    currentframe += 1
                else:
                    break

            # Release all space and windows once done
            cam.release()
            cv2.destroyAllWindows()

            

create_frames_from_videos()
```
